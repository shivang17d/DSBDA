```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("cancer.csv")
```

```python
df
```

```python
df.drop("Unnamed: 32",axis='columns',inplace=True)
```

```python
df
```

```python
colors = ['hotpink','orange']
test = ['M','B']
```

```python
for i in range(2):
    x = df[df['diagnosis'] == test[i]]
    plt.scatter(x['radius_mean'],x['texture_mean'],c=colors[i],label=test[i])
plt.xlabel("radius_mean")
plt.ylabel("smoothness_mean")
plt.legend()
plt.show()
```

```python
for i in range(2):
    x = df[df['diagnosis'] == test[i]]
    plt.scatter(x['area_mean'],x['texture_mean'],c=colors[i],label=test[i])
plt.xlabel("area_mean")
plt.ylabel("texture_mean")
plt.legend()
plt.show()
```

```python
plt.pie(df['diagnosis'].value_counts(),labels = ['M','B'])
plt.title("Diagnosis")
```

```python
df['diagnosis'] = df['diagnosis'].map({'M':0,'B':1})
df
```
